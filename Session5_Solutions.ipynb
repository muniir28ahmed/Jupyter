Create a lazy iterable with a list of animals and use a for loop to iterate over it.

```python
def filter_words(words, X):
    for w in words:
        if len(w) < X:
            yield w

res = filter_words(['cat','elephant','dog', 'girafe'],X=5)
```

Write a function that returns true if the string passed to its single argument is a palindrome (same forwards as backwards).



```python
def palin(s):
    return str(s) == str(s)[::-1]
```

Use list comprehension, and the two functions you have just created, to pick only the numbers in the first 10000 numbers of the Fibonacci series that are palindromes.

```python
f = fibbo(10000)
p = [n for n in f if palin(n)]
print(p)
```

## 5.7 Recursion

Write a recursive function to print the Fibonacci sequence

```python
def recursive_fibo(n):
   if n <= 1:
       return n
   else:
       return(recursive_fibo(n-1) + recursive_fibo(n-2))
```

Now call it and print the 10th number in the sequence.

```python
recursive_fibo(10)
```

Try another recursive function - create one to calculate factorials (remember that n factorial is the product of the whole numbers between 1 and n).

```python
def calculateFactorial(n):
    if n == 1:
        return 1
    else:
        return n*calculateFactorial(n-1)
    
calculateFactorial(10)
```

Factorials are interesting because of how quickly the number grows - use a loop to print the factorials between 1 and 20

```python
for i in range(1,20):
    print(calculateFactorial(i))
```
