Write a generator function that receives 2 arguments: a list of words and an integer called X. This function should iterate over the list of words and only return those with fewer letters than X. 

Create a lazy iterable with a list of animals and use a for loop to iterate over it.

```python
def m(animal, X):
    for i in animal:
        if len(i) < X:
            yield i
            
one = m(['muniir','hi', 'osman', 'u'],6)
for a in one:
    print(a)
```

Write a generator function to generate the first n numbers in the Fibonacci series. For a brief explanation of the Fibonacci series, please see https://science.howstuffworks.com/math-concepts/fibonacci-nature.htm 

```python
def fibbo(n):
    yield 1
    yield 1
    a=1
    b=1
    for i in range(n-2):
        x = a+b
        yield x
        a=b
        b=x        
```

The function should be created like this:

```python
f = fibbo(10000)
```

And called like this:

```python
next(f)
```

Write a function that returns true if the string passed to its single argument is a palindrome (same forwards as backwards).



```python
def is_palindrome(s):
    return str(s) == str(s)[::-1]
is_palindrome(10000)
```

Use list comprehension, and the two functions you have just created, to pick only the numbers in the first 10000 numbers of the Fibonacci series that are palindromes.

```python
f = fibbo(10000)
m =[n for n in f if is_palindrome(n)]
print(m)
```

## 5.7 Recursion

Write a recursive function to print the Fibonacci sequence

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return (fibonacci(n-1) + fibonacci(n-2))
```

Now call it and print the 10th number in the sequence.

```python
fibonacci(10)
```

Try going a little higher, the 20th number in the sequence for example.

Try another recursive function - create one to calculate factorials (remember that n factorial is the product of the whole numbers between 1 and n).

```python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n*factorial(n-1)
factorial(10)
```

Factorials are interesting because of how quickly the number grows - use a loop to print the factorials between 1 and 20

```python
n = 20
i = 1
while i <= n:
    print(factorial(i))
    i += 1
```
