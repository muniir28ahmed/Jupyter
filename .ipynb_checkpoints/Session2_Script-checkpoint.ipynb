## 2.7 RMD22 Data Types
I've talked about variables containing data, but the example so far has stored only single numbers. You will learn later about variables that can store a wealth of different types of data, but now we will look at the different simple types of data that a variable can store.

First, we make the distinction between numbers and strings. You have already see how string literals have quotation marks around them so that Python knows they are strings. We can illustrate the difference in a simple example:

# Data Types

```python
years = 10
print(years*2)
years = "10"
print(years*2)
```

Here we created a variable called years and assigned to it the value 10. Python took this to be a number so when we multiplied it by 2, we got the answer 20 as expected. In the next row, we assigned the string value one-oh to our variable called years. The quotation marks indicate it is a string. Now when we print years times 2, we see the string one-oh one-oh - which is the years string printed twice. So in Python, the operations you perform on your variables can depend on the data types the variables store.

The basic data types in Python are strings, which we have already met, integers, which are whole numbers like 10, floating point numbers, known as floats, which have a decimal point, for example, 3.5 or 0.1, and Boolean values which can be either true or false.

Python tries to work out what type of data you are storing in a variable and acts on it accordingly. So if I say years equals 10, then years will be an integer but if I say years equals 10 point 5 then it will be a float. Wrapping a literal value in quotation marks makes it a string. Each of these types has a name that Python uses to identify it. For strings, it is str, that is s-t-r, for integers, it is int, for floating point numbers it is float, and for Boolean values, it is bool. Python has a function called type, which takes one argument and tells you the data type of whatever you pass to it. Python also has one function for each of the data types I just listed, which converts the value passed to it into its named type. For example, the function called str converts whatever you pass to it into a string.

Let's make use of those functions now. First we try a few different values and variables in the type function. Remember that all functions take an argument list that is encloded in round brackets. We only want to pass one argument to the function, so we enclose that in round brackets. Let's try with an integer first. We can pass it a literal or a variable.

```python
print(type(1))
a=1
print(type(a))
```

See how both the literal value one and a variable that contains the value one report their type as being int. Next we try strings in the same way. See how the type is reported as being str.

```python
print(type("Hello"))
a="Hello"
print(type(a))
```

Now we try it with a floating point number. That is recognised as a float

```python
print(type(1.5))
```

and finally the values true and false are recognised as Booleans. Notice how the words true and false do not have quotation marks around them. If they did, they would be strings. We can try that too.

```python
print(type(False))
print(type(True))
```

```python
print(type("False"))
print(type("True"))
```

Now we will use the functions that share their names with the data types to convert data to the type we need. First, lets convert an integer to a string. When I run this code, it will create a variable called a, which contains the integer value 5. It will then create a variable called b, which is a string representation of the value in a.

```python
a = 5
b = str(a)
print(type(a))
print(type(b))
```

Now it has run, you can see the data types being reported. a is an int and b is a string. Why would we bother doing that? Well, remember earlier we saw how multiplying a value by two had a different effect depending on the data type? If we have data in a variable that is a string, and we want to perform a numeric operation on it, like doubling it, then we need to change it into a number first. This could happen if the data are read as strings from a file or, as we will do in the next example, if the data are entered by a user.

If you want to allow a user to enter data for a Python script to process, a simple way to do so is with an input function. This function causes the program to wait for the user to type something in, and then makes whatever they typed available to the program via a variable. Here is an example:

```python
print("Enter a number and I will calculate its square")
num = input()
print("The square of", num, "is", num*num)
```

There we have our first Python error! You will see more of these as you learn to program. Python is telling me that it cannot perform the calculation I have asked for, which is multiplying a string by a string. We can fix that using the int function to change the string to an integer. Notice how it is pointing out which line in our code caused the problem.

```python
print("Enter a number and I will calculate its square")
num = input()
intnum = int(num)
print("The square of", num, "is", intnum*intnum)
```

There we created a new variable called intnum and used the function called int to convert the string called num to an integer. Remember, the variable names mean nothing to Python, so I could have called intnum anything I wanted. The fact that I called it intnum is not why it is now treated as an integer. That is done by calling the function called int.

This would be a good time to point out something else about functions. A few functions, like print, for example, are used to tell Python to perform some operation - printing to the screen in that case. Most, however, are used to perform some kind of calculation and to provide the result of doing so. The other functions we have used so far, including type and int are like that. We say that they return a value. The function called int returns a value that is the integer representation of the argument that you pass to it. The function called type returns something that describes the type of the argument you pass to it. So functions accept arguments and return a result.

## RMD23 2.8 Mutability

The simple data types that we have introduced so far store a single value in memory an allow you to access that value by the name you first gave it. Every time you assign a value to a variable of a simple data type like that, it uses a new location in memory, regardless of whether or not you have a variable with that name already. For example, if we start with a variable called age and give it the value 10 and we then want to add one to it then we can type this

# Mutability

```python
age = 10 
print(age)
age = age+1
print(age)
```

Behind the scenes, Python grabs a new piece of memory and stores the value 10 there when the first line is run. Then, when it gets to the second line, it grabs another new piece of memory and stores the value 11 there. It no longer uses the first value, where age was 10 - that is discarded, so when we use the variable called age, we get the value 11. We describe this quality whereby variables are not overwritten, but are instead assigned in a new part of memory by saying a variable is immutable. Right now, none of the consequences of this will be obvious, but as we go on through the course, they will start to make themselves felt, so we should start now but at least mentioning the idea that simple data types are immutable. Later we will meet data types that are mutable (they are changed in situ) and start to understand why this distinction is important.

Just now, I will show you one thing that you will need to know about to understand mutability - the function called ID. Pass a variable to the id function and it will give you an identifier for where the data associated with that variable are stored. A quick demonstration will start the ball rolling:

```python
age = 10
print(id(age))
age = age+1
print(id(age))
```

See how the memory id for age when I first defined it is different from the location after I have added 1 to it. The second use of the variable called age is actually a brand new variable that happens to share a name with the old variable called age, which is no longer available to us. I imagine you are wondering why it matters. A lot of the time, it doesn't, but occasionally, it does, and so you need to know what we mean when we talk about mutability later.
