## Calculate a volume

# f-strings
f-string syntax is `f"some text {variable} more text"`. Repeat the kilogram calculation code you used above, but this time use f-strings to print the result.

```python
print("Enter a weight in kilograms and I will calculate its weight in pounds")
kilos = input()
pounds = float(kilos)*2.2
print(f"{kilos} kilograms is {pounds} pounds")
```
